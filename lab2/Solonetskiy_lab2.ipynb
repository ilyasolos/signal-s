### Імпортуємо необхдіні пакети та бібліотеки

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Завдання
## 1. Порівняти гістограми двох зображень.

1.1. Відкрити та завантажити два зображення:
 

```python
img_half = cv2.imread("image\input\Lab3_half100.png", cv2.IMREAD_GRAYSCALE)
img_chess = cv2.imread("image\input\Lab3_chess100.png", cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(8, 3))
plt.subplot(121), plt.imshow(img_half, cmap = 'gray'), plt.title('half'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(img_chess, cmap = 'gray'), plt.title('сhess'), plt.xticks([]), plt.yticks([])

plt.show();
```

1.2.Визначити гістограму для кожного зображення:

```python
hist_half = cv2.calcHist([img_half], [0], None, [256], [0, 256])
hist_chess = cv2.calcHist([img_chess], [0], None, [256], [0, 256])
```

1.3.Відобразити завантаженні зображення та відповідно визначені гістограми зображень за допомогою функції **plt.subplot**:

```python
plt.figure(figsize=(8, 3))
plt.subplot(121), plt.imshow(img_half, cmap = 'gray'), plt.title('half'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.plot(hist_half), plt.title('histogram-half')

plt.figure(figsize=(8, 3))
plt.subplot(121), plt.imshow(img_chess, cmap = 'gray'), plt.title('chess'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.plot(hist_chess), plt.title('histogram-chess')


plt.show()
```
