## Додаткові завдання:

#### 1. Перетворити зображення з колірної моделі BGR в колірну модель HSV за допомогою формул.

```python
img_brg = cv2.imread("input\image.jpg")


def bgr_to_hsv(b, g, r):                   
    r, g, b = r/255.0, g/255.0, b/255.0
    mx = max(r, g, b)
    mn = min(r, g, b)
    df = mx-mn
    if mx == mn:
        h = 0
    elif mx == r:
        h = (30 * ((g-b)/df) + 180) % 180
    elif mx == g:
        h = (30 * ((b-r)/df) + 60) % 180
    elif mx == b:
        h = (30 * ((r-g)/df) + 120) % 180
    if mx == 0:
        s = 0
    else:
        s = (df/mx)*255
    v = mx*255
    return int(round(h)), int(round(s)), int(round(v))


img_hsv = []
for i in range(img_brg.shape[0]):
    for j in range(img_brg.shape[1]):
        img_hsv.append([bgr_to_hsv(img_brg[i,j,0],img_brg[i,j,1],img_brg[i,j,2])])

img_hsv = np.array(img_hsv)
img_hsv.shape = (img_brg.shape[0],img_brg.shape[1],3)    

plt.figure(figsize=(8, 3))
plt.subplot(121), plt.imshow(img_brg), no_axes('BGR')
plt.subplot(122), plt.imshow(img_hsv), no_axes('HSV')

plt.show()

cv2.imwrite("output\img_brg_dop.png", img_brg)
cv2.imwrite("output\img_hsv_dop.png", img_hsv)
```

```python
A0 = ((233, 45, 13), #  [X0 Y0 Z0]
      (14, 5, 133))  #  [X1 Y1 Z1] 

L = (A0[0][0], A0[1][0])
a = (A0[0][1], A0[1][1])
b = (A0[0][2], A0[1][2])

C1 = pow( pow(a[0],2) + pow(b[0],2), 1/2)
C2 = pow( pow(a[1],2) + pow(b[1],2), 1/2)

KL, KC, KH = 1, 1, 1
K1,K2 = 0.045, 0.015

SL = 1
SC = 1 + K1 * C1
SH = 1 + K2 * C2

delta_L = A0[0][0] - A0[1][0]
delta_a = A0[0][1] - A0[1][1]
delta_b = A0[0][2] - A0[1][2]
delta_C = C1 - C2
delta_H = pow( pow(delta_a,2) + pow(delta_b,2) + pow(delta_C,2) ,1/2)

#http://www.brucelindbloom.com/index.html?Eqn_DeltaE_CIE94.html
#https://en.wikipedia.org/wiki/Color_difference#cite_note-14

E74 = pow(pow(delta_L,2)+pow(delta_a,2)+pow(delta_b,2),1/2)
E94 = pow(pow(delta_L/(KL*SL),2)+pow(delta_C/(KC*SC),2)+pow(delta_H/(KH*SH),2),1/2)

print("delta E74 = " + str(E74))
print("delta E94 = " + str(E94))
```
