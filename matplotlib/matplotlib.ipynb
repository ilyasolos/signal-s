# Завдання
## 1. Створити Subplot із  4 графіків сінусоїди з різними частотами

1.1. Сформувати діапазон функції від 0 до 2 з кроком 0.01

```python
t = np.arange(0, 2, 0.01)
```

1.2 Сформувати частоти сінусоїди

```python
s1 = np.sin(4 * np.pi * t)
s2 = np.sin(6 * np.pi * t)
s3 = np.sin(8 * np.pi * t)
s4 = np.sin(10 * np.pi * t)
```

1.3 Надрукувати графіки

```python
plt.plot(t, s1, c='r',linewidth=2.0)
plt.plot(t, s2, c='b',linewidth=2.0)
plt.plot(t, s3, c='g',linewidth=2.0)
plt.plot(t, s4, c='black',linewidth=2.0)
plt.grid()
plt.show()
```

1.2 Сформувати частоти сінусоїди

1.2 Сформувати частоти сінусоїди

```python
fig = plt.figure(figsize=(15,15))
ax0 = fig.add_subplot(2, 2, 1)
ax0.plot(t, s1, c='g',linewidth=2.0)
ax0.set(xlabel='x', ylabel='y', title='Сінус функція №1')
ax0.grid()

ax1 = fig.add_subplot(2, 2, 2)
ax1.plot(t, s2, c='g',linewidth=2.0)
ax1.set(xlabel='x', ylabel='y', title='Сінус функція №2')
ax1.grid()

ax2 = fig.add_subplot(2, 2, 3)
ax2.plot(t, s3, c='black',linewidth=2.0)
ax2.set(xlabel='x', ylabel='y', title='Сінус функція №3')
ax2.grid()

ax3 = fig.add_subplot(2, 2, 4)
ax3.plot(t, s4, c='black',linewidth=2.0)
ax3.set(xlabel='x', ylabel='y', title='Сінус функція №4')
ax3.grid()

plt.show()
```
