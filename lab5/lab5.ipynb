### Імпортуємо необхдіні пакети та бібліотеки

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def foto(img, name):
    return plt.imshow(img, cmap='gray'), plt.xticks([]), plt.yticks([]), plt.title(str(name))
```

# Завдання
## 1. Методами морфологічної обробки збільшити та зменшити контури зображення
1.1 Завантажити та відобразити зображення "1_8.png" директорії "./examples/"

```python
img = cv2.imread("input/1_8.png", cv2.IMREAD_GRAYSCALE)
```

```python
plt.figure(figsize=(4, 4))
foto(img, "baza")
plt.show();
```

1.2 Створити бінарне зображення (зображення із двома значеннями яскравості пікселів - мінімальним та максимальним) методом порогової обробки. Порогове значення яскравості встановити 10.

```python
img_bin = img.copy()
threshold = 10
max_val = 255
min_val = 0
```

```python
for row in range(img.shape[0]):
    for col in range(img.shape[1]):
        if (img[row][col]) > threshold:
            img_bin[row][col] = max_val
        else:
            img_bin[row][col] = min_val
```

1.3 Відобразити бінарне зображення

```python
plt.figure(figsize=(5, 5))
foto(img_bin, "binar baza")
plt.show();
```

1.4 Встановити ядро розміром 2х2

```python
n = ([2,2])
kernel = np.ones(n, np.uint8)
```

1.5 Застосувати ерозію до бінарного зображення та відобразити його

```python
erosion_binar = cv2.erode(img_bin, kernel)

plt.figure(figsize=(8, 5))
foto(erosion_binar, "binar erosion baza")
plt.show();
```

1.6 Застосувати діляцію до бінарного зображення та відобразити його

```python
dilation_binar = cv2.dilate(erosion_binar, kernel)

plt.figure(figsize=(8, 5))
foto(dilation_binar, "dilation binar baza")
plt.show();
```

## 2. Методами морфологічної обробки підготувати зображення до подальшої обробки
2.1 Завантажити та відобразити зображення "street.jpg" директорії "./examples/"

```python
baza = cv2.imread("input\street.jpg", cv2.IMREAD_GRAYSCALE)
```

2.2 Створити бінарне зображення (зображення із двома значеннями яскравості пікселів - мінімальне та максимальне) методом порогової обробки. Порогове значення яскравості встановити 170.

```python
baza_bin = baza.copy()
threshold = 165
max_val = 255
min_val = 0
```

```python
for row in range(baza.shape[0]):
    for col in range(baza.shape[1]):
        if (baza[row][col]) > threshold:
            baza_bin[row][col] = max_val
        else:
            baza_bin[row][col] = min_val
            
n = ([2,2])
kernel = np.ones(n, np.uint8)
```

2.3 Відобразити бінарне зображення

```python
plt.figure(figsize=(8, 8))
foto(baza_bin, "binar baza")
plt.show();
```

2.4 Встановити ядро розміром 2х2

```python
n = ([2,2])
kernel = np.ones(n, np.uint8)
```

2.5 Застосувати ерозію до бінарного зображення та відобразити його

```python
erosion = cv2.erode(baza_bin,kernel)

plt.figure(figsize=(8, 5))
foto(erosion, "erosion baza")
plt.show();
```

2.6 Застосувати морфологічне відкриття до зображження після обробки ерозією та відобразити його

```python
opening = cv2.morphologyEx(erosion, cv2.MORPH_GRADIENT, kernel)

plt.figure(figsize=(8, 5))
foto(opening, "opening baza")
plt.show();
```
